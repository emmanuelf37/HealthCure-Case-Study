# #3 Creating Matrix Heatmap:

```python
# Create a correlation matrix using pandas

df.corr()
```

```python
df.columns
```

```python
# Create a subset excluding "Start Date", "End Date", "Group"

sub = df[['State', 'Sex',
       'Age Group', 'COVID-19 Deaths', 'Total Deaths', 'Pneumonia Deaths',
       'Pneumonia and COVID-19 Deaths', 'Influenza Deaths',
       'Pneumonia, Influenza, or COVID-19 Deaths']]
```

```python
sub
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), cmap="YlGnBu", annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap. 
```

This chart leaves out the groupings becuase they are mostly made of blanks and are meant to be filtered using excel, think. And the other categories were left out becuase they were non-numerical. This leaves only the different death counts. Several things we can gleam from this are which diseases have the highest correlation to the "Total Deaths" meaning the more deaths in that disease category, the larger the total death count tends to be. This disease would be Pneumonia and Pneumonia+ COVID-19. I think most of the others correllations between diseases are just because they are all happening around the same time.
